```python

cmoa_name=cmoa_data2[:,5]
#print(cmoa_data2[:,0])
print(cmoa_name)
```

```python
# Firstly, use the for function and count function to count the '|' of occurrence
# Secondly, the number of artists = the number of '|' +1
name_numbers=[]
for i in range (len(cmoa_name)):
    name=cmoa_name[i]
    #print(name.count('|'))
    name_number=name.count('|')+1
    name_numbers.append(name_number)
    print(name_numbers)
```

```python
# This coding can be used to see the number of artisits in each line 
for i in range (len(cmoa_name)):
    name2=cmoa_name[i]
    #print(name.count('|'))
    name_number2=name2.count('|')+1
    print(name_number2)
```

```python
# Use the mode function to find the same number of artisits and all different group size
from collections import Counter
mode=Counter(name_numbers).most_common(1000000000)
print(mode)
# so in the each(), the first one is the group sizes and the second one is the artefact which is made by this group size
```
